# Regress√£o linear do zero

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = torch.rand(100, 1)
y = X * 3 + 12
```

```python
plt.plot(X, y, 'r-')
plt.show()
```

```python
def predict(X, a, b):
    return (X * a) + b

def loss(y_pred, y):
    return torch.mean((y_pred - y) ** 2)
```

```python
a = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)
learning_rate = 0.01

for i in range(2000):
    y_pred = predict(X, a, b)
    current_loss = loss(y_pred, y)

    current_loss.backward()
    with torch.no_grad():
        a -= learning_rate * a.grad
        b -= learning_rate * b.grad
    a.grad.zero_()
    b.grad.zero_()

display(a, b)
```
